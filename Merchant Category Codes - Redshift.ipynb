```python
#Get connection Redshift
def redshift_conn():
    config = {}
    config["redshift_user"] = xxxxxx
    config["redshift_password"] = xxxxxx
    config["redshift_host"] = xxxxxx
    config["redshift_port"] = xxxxxx
    config["redshift_database"] = xxxxxx
    uri = 'redshift+psycopg2://{p[redshift_user]}:{p[redshift_password]}@{p[redshift_host]}:{p[redshift_port]}/{p[redshift_database]}'\
    .format(p=config)
    db_credentials = (uri)
    engine = sa.create_engine(db_credentials)
    Session = sessionmaker(bind=engine)
    return Session(), uri

conn, uri = redshift_conn()
```
